```python
!pip install transformers
!pip install sentence_transformers
!pip install requests
!pip install bs4
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup #BeautifulSoup is a web scrapping tool used to parse html and xml documents
from transformers import pipeline #used to generate text using a pre-trained model
from sentence_transformers import SentenceTransformer #used to convert text into vectors
from sklearn.metrics.pairwise import cosine_similarity #used to calculate the similarity between two vectors
```

```python
!huggingface-cli login ##accessing the HuggingFace platform using the account token
```

```python
#accessing openAI's gpt2 model for text generation from the HuggingFace

generator_gpt2 = pipeline('text-generation', model='openai-community/gpt2') #Hugging Face model hub is a centralized repository of pre-trained models including openai
```

```python
##loading a SentenceTransformer model (all-MiniLM-L6-v2) for embeddings from the HuggingFace
#generating semantically meaningful embeddings for sentences
#word embeddings are most common method of vectorization
embedding_model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
url = 'https://kindxiaoming.github.io/pykan/intro.html'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser') #creates a parsed representation of the HTML content.
```

```python
soup
```

```python
#in html p->paragraph,h1->header 1,h2->header 2,h3->header 3
text_chunks = [tag.get_text().strip() for tag in soup.find_all(['p', 'h1', 'h2', 'h3']) if tag.get_text().strip()]
print(text_chunks)
```

```python
#converting to a dataframe
df_text_chunks = pd.DataFrame(text_chunks)
df_text_chunks
```

RAG: Indexing

```python
# word embedding process
document_embeddings = embedding_model.encode(text_chunks)
print(document_embeddings)
```

```python
#Converting to a dataframe
df_document_embeddings = pd.DataFrame(document_embeddings)
df_document_embeddings
```

```python
#joining two dataframes (chunks and associated embeddings)

df_chunks_embeddings = pd.concat([df_text_chunks, df_document_embeddings], axis=1)
df_chunks_embeddings
```

```python
#reloading our question

question = "Give me the code to plot KAN at initialization"
```

```python
#embedding the question

question_embedding = embedding_model.encode([question])[0]
print(question_embedding)
```

RAG: Retrieval


```python
#retrieve the most relevant documents based on cosine similarity between question and text chunks
similarities = cosine_similarity([question_embedding], document_embeddings)[0]
similarities
```

```python
#getting top four documents
top_indices = np.argsort(similarities)[::-1][:4]
top_indices
```

```python
#checking the selected documents/chunks/paraghraphs using the top four array index numbers.
#search for the array indices here for better understanding
df_chunks_embeddings
```

```python
#top-k documents based on similarities between question and document embedding distances
top_k_doc = [(text_chunks[i], similarities[i]) for i in top_indices]
top_k_doc
```

RAG: Response Generation


```python
#combining the retrieved documents to form a context
context = " ".join([doc for doc, _ in top_k_doc])
print(context)
```

```python
#passing system prompt
system_prompt = "Please provide a concise and accurate response."
system_prompt
```

```python
#combining system prompt with question and context
prompt = f"system: {system_prompt} Question: {question} Context: {context}"
```

LLM response after the RAG architecture integration

```python
#generating the response using GPT-2 model

#response = generator_gpt2(prompt, max_new_tokens=50, num_return_sequences=1)[0]['generated_text']
response_with_rag = generator_gpt2(prompt, max_new_tokens=50, num_return_sequences=1)[0]['generated_text']
```

```python
#print(response,end="/n")
print("Response with RAG:\n", response_with_rag, end="\n\n")
```

```python
!pip install evaluate

from evaluate import load
import evaluate

# Load the ROUGE metric
rouge = evaluate.load('rouge')

# Use the generated response as the candidate
candidates = [response_with_rag]

# Define your references
references = [
    [
        "Kolmogorov-Arnold representation theorem states that if 𝑓 is a multivariate continuous function on a bounded domain, then it can be written as a finite composition of continuous functions of a single variable and the binary operation of addition.",
        "In a sense, they showed that the only true multivariate function is addition, since every other function can be written using univariate functions and sum."
    ]
]

# Compute ROUGE scores
# ROUGE-1 measures the overlap of unigrams between the candidate and reference summaries.
# ROUGE-2 measures the overlap of bigrams between the candidate and reference summaries.
# ROUGE-L measures the longest common subsequence (LCS) between the candidate and reference summaries.
# ROUGE-Lsum measures the LCS at the summary level rather than the sentence level.
results = rouge.compute(predictions=candidates, references=references)
print(results)
```
